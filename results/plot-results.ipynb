```julia
using CSV
using DataFrames
using Plots
using StatsPlots
```

```julia
julia_n2plain = CSV.read("julia-n2plain.csv", DataFrame)
```

```julia
julia_n2tiled = CSV.read("julia-n2tiled.csv", DataFrame)
```

```julia
@df julia_n2plain plot(:mean_particles, :time_per_event, label = "Julia N2Plain")
@df julia_n2plain scatter!(:mean_particles, :time_per_event, label = "Julia N2Plain")
@df julia_n2tiled plot!(:mean_particles, :time_per_event, label = "Julia N2Tiled")
@df julia_n2tiled scatter!(:mean_particles, :time_per_event, label = "Julia N2Tiled")
```

```julia
@df julia_n2tiled plot(:mean_particles, :time_per_event, label = "Julia N2Tiled")
@df julia_n2tiled scatter!(:mean_particles, :time_per_event, label = "Julia N2Tiled")
```
