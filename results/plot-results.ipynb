# Jet Reconstruction Results Plotter

Notebook for making plot results from jet reconstruction benchmarks

```julia
results_dir = "OSX-14.5-M2Pro-Julia-1.10.4"
algorithm = "AntiKt"
radius = "0.4"
```

```julia
using CSV
using DataFrames
using Plots
using StatsPlots
```

```julia
julia_n2plain = CSV.read(joinpath(results_dir, "Julia-$(algorithm)-N2Plain-$(radius).csv"), DataFrame);
```

```julia
julia_n2tiled = CSV.read(joinpath(results_dir, "Julia-$(algorithm)-N2Tiled-$(radius).csv"), DataFrame);
```

```julia
@df julia_n2plain plot(:mean_particles, :time_per_event, label="", xlabel="<n> initial particles", ylabel="Jet Reconstruction Time μs/event")
@df julia_n2plain scatter!(:mean_particles, :time_per_event, label = "Julia N2Plain")
@df julia_n2tiled plot!(:mean_particles, :time_per_event, label="")
@df julia_n2tiled scatter!(:mean_particles, :time_per_event, label = "Julia N2Tiled")
```

```julia
savefig(joinpath(results_dir, "julia-jet-reco.pdf"))
```

```julia
fastjet_n2plain = CSV.read(joinpath(results_dir, "FastJet-$(algorithm)-N2Plain-$(radius).csv"), DataFrame);
```

```julia
fastjet_n2tiled = CSV.read(joinpath(results_dir, "FastJet-$(algorithm)-N2Tiled-$(radius).csv"), DataFrame);
```

```julia
@df julia_n2plain plot(:mean_particles, :time_per_event, label="", xlabel="<n> initial particles", ylabel="Jet Reconstruction Time μs/event")
@df julia_n2plain scatter!(:mean_particles, :time_per_event, label = "Julia N2Plain")
@df fastjet_n2plain plot!(:mean_particles, :time_per_event, label = "")
@df fastjet_n2plain scatter!(:mean_particles, :time_per_event, label = "FastJet N2Plain")
```

```julia
savefig(joinpath(results_dir, "julia-fastjet-reco-n2plain.pdf"))
```

```julia
@df julia_n2tiled plot(:mean_particles, :time_per_event, label="", xlabel="<n> initial particles", ylabel="Jet Reconstruction Time μs/event")
@df julia_n2tiled scatter!(:mean_particles, :time_per_event, label = "Julia N2Tiled")
@df fastjet_n2tiled plot!(:mean_particles, :time_per_event, label = "")
@df fastjet_n2tiled scatter!(:mean_particles, :time_per_event, label = "FastJet N2Tiled")
```

```julia
savefig(joinpath(results_dir, "julia-fastjet-reco-n2tiled.pdf"))
```
